#### **1-4.** 다음의 조건을 만족하는 새로운 개체(DataFrame) ‘CL’ 및 ‘TR’을 만드시오.
● CL : 클린턴 후보의 자료, TR : 트럼프 후보의 자료 <br>
● 50개 주 각각에서 후보의 실제 득표율, 예측 득표율, 표본 수를 저장 <br>
● 예측 득표율의 경우, 선거일 기준 가장 최근의 날짜에 집계된 결과만 활용(누적된 결과이므로) <br>
● 총 열 개수는 4개가 되어야 함 - 열 이름 : State, Real, Prediction, Sample_size <br>
● 총 행 개수는 워싱턴 D.C.를 포함하여 51개가 되어야 함 <br>

```python
# 빈 데이터프레임 만들기
columns2 = ['State','Real','Prediction','Sample_size']
CL = pd.DataFrame(columns=columns2)
CL
```

```python
# 주 이름 모음
statelist2 = np.unique(polls['state'])
len(statelist2)
```

```python
# 각 주별로 선거일과 가장 가까운 날에 집계된 결과만 정리
B = polls.loc[polls.groupby(['state'])['days_to_election'].idxmin()]
B = B.reset_index()
B.head()
```

```python
# Clinton 후보
CL['State'] = statelist2
# CL['Real'] = 순서정리 필요
CL['Prediction'] = B['Clinton']
CL['Sample_size'] = B['sample_size']
CL.head()
```

results 표와 CL, TR 표의 주 이름 순서가 달라서, CL, TR 표의 주 이름 순서에 맞도록 하여 adjusted라는 list에 저장했습니다.

```python
adjusted = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'DC', 'FL', 'GA',
       'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA',
       'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY',
       'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
       'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'] 
```

```python
# CL['Real'] 순서정리 및 실제 득표율 모음
for i in range(51):
    S = B['state'][i]
    for j in range(51) :
        D = statelist2[j]
        if  S == D :
            CL['Real'][i] = results[results['State']==adjusted[j]]['Clinton'].values[0]
```

```python
# 결과 확인
CL.head()
```

```python
# 빈 데이터프레임 만들기
TR = pd.DataFrame(columns=columns2)
TR
```

```python
# Trump 후보
TR['State'] = statelist2
# TR['Real'] = 순서정리 필요
TR['Prediction'] = B['Trump']
TR['Sample_size'] = B['sample_size']
TR.head()
```

```python
# TR['Real'] 순서정리 및 실제 득표율 모음
for i in range(51):
    S = B['state'][i]
    for j in range(51) :
        D = statelist2[j]
        if  S == D :
            TR['Real'][i] = results[results['State']==adjusted[j]]['Trump'].values[0]
```

```python
# 결과 확인
TR.head()
```

#### **1-5.** 후보별 득표율의 예측오차 분포를 히스토그램으로 시각화하시오.

##### 1) Clinton

```python
# Bias_C : Clinton 후보의 예측오차
Bias_C = abs(CL['Real'] - CL['Prediction'])

# Histogram
CL_prederror_hist = plt.subplot()
CL_prederror_hist.hist(Bias_C)
CL_prederror_hist.set_xlabel("Prediction Error")
CL_prederror_hist.set_ylabel("Frequency")
CL_prederror_hist.set_title("Histogram of Prediction Error of Clinton's Polling Rate")
```

##### 2) Trump

```python
# Bias_T : Trump 후보의 예측오차
Bias_T = abs(TR['Real'] - TR['Prediction'])

# Histogram
TR_prederror_hist = plt.subplot()
TR_prederror_hist.hist(Bias_T)
TR_prederror_hist.set_xlabel("Prediction Error")
TR_prederror_hist.set_ylabel("Frequency")
TR_prederror_hist.set_title("Histogram of Prediction Error of Trump's Polling Rate")
```

#### **1-6.** ‘CL’과 ‘TR’에 각각 표준오차를 담은 ‘SE’라는 열(column)을 추가하시오.

```python
CL_SE = np.sqrt(CL['Prediction'] * (1 - CL['Prediction']) / CL['Sample_size'])
CL['SE'] = CL_SE
CL.head()
```

```python
TR_SE = np.sqrt(TR['Prediction'] * (1 - TR['Prediction']) / TR['Sample_size'])
TR['SE'] = TR_SE
TR.head()
```

#### **1-7.** ‘CL’과 ‘TR’에 각각 근사신뢰구간(approximate confidence interval)의 하한과 상한을 나타내는 열(column) ‘lower’과 ‘upper’를 추가하시오(유의수준 5%).

```python
# 분위수값 구하는 방법
from scipy.stats import norm
Z = norm(loc = 0, scale = 1) # 표준정규분포를 따르는 확률변수 만들기
Z.ppf(0.95)
```

```python
CL_lower = CL['Prediction'] - Z.ppf(0.95) * CL['SE']
CL_upper = CL['Prediction'] + Z.ppf(0.95) * CL['SE']
CL = CL.assign(lower = CL_lower, upper = CL_upper)
CL.head()
```

```python
TR_lower = TR['Prediction'] - Z.ppf(0.95) * TR['SE']
TR_upper = TR['Prediction'] + Z.ppf(0.95) * TR['SE']
TR = TR.assign(lower = TR_lower, upper = TR_upper)
TR.head()
```

#### **1-8.** 클린턴 후보와 트럼프 후보 각각의 여론조사 신뢰수준을 구하시오.

##### 1) Clinton

```python
CL_include = (CL['lower'] <= CL['Real']) & (CL['Real'] <= CL['upper'])
print("Clinton's Polling Rate Confidence Level:", round(sum(CL_include) / len(CL_include) * 100, 4), "%")
```

##### 2) Trump

```python
TR_include = (TR['lower'] <= TR['Real']) & (TR['Real'] <= TR['upper'])
print("Trump's Polling Rate Confidence Level:", round(sum(TR_include) / len(TR_include) * 100, 4), "%")
```

Clinton을 지지하는 사람들이 Trump를 지지하는 사람들보다 여론조사에 참여하는 비율이 높았고, 이는 Clinton의 지지율이 과대평가되는 결과로 이어졌다. 이로인해 Clinton의 득표율 여론조사 신뢰수준이 Trump보다 높았고, 이는 실제 선거 결과와 반대되는 여론조사 결과로 이어진 것이다.
